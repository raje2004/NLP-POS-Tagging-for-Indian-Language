```python
def parse_conllu(file_path):
    sentences = []

    with open(file_path, 'r', encoding='utf-8') as file:
        sentence = []

        for line in file:
            # Skip comments and empty lines
            if line.startswith('#') or not line.strip():
                if sentence:
                    sentences.append(sentence)
                    sentence = []
                continue

            parts = line.split('\t')

            # CoNLL-U format = 10 columns
            if len(parts) != 10:
                continue

            word = parts[1]        # FORM
            pos_tag = parts[3]     # UPOS tag

            sentence.append((word, pos_tag))

        if sentence:
            sentences.append(sentence)

    return sentences
```

```python
train_file = "hi_hdtb-ud-train.conllu"
test_file = "hi_hdtb-ud-test.conllu"

print("Loading dataset...")
```

```python
train_sents = parse_conllu(train_file)
test_sents = parse_conllu(test_file)

print(f"Train sentences: {len(train_sents)}")
print(f"Test sentences: {len(test_sents)}")
```

```python
print("\nTraining taggers...")

default_tagger = DefaultTagger('NOUN')

unigram_tagger = UnigramTagger(train_sents, backoff=default_tagger)

bigram_tagger = BigramTagger(train_sents, backoff=unigram_tagger)

trigram_tagger = TrigramTagger(train_sents, backoff=bigram_tagger)
```

```python
print("\nEvaluating model...")

accuracy = trigram_tagger.accuracy(test_sents)

print(f"\nModel Accuracy: {accuracy:.2%}")
```

```python
def tag_sentence(sentence):
    tokens = trivial_tokenize(sentence, lang='hi')
    tagged = trigram_tagger.tag(tokens)
    return tagged
```

```python
print("\n--- Sample Predictions ---")

samples = [
    "दो आदमी आए।",
    "राम स्कूल जाता है।",
    "भारत एक महान देश है।"
]

for sent in samples:
    tagged_output = tag_sentence(sent)
    print("\nInput:", sent)
    print("Output:", " ".join([f"{w}/{t}" for w, t in tagged_output]))

```

```python
print("\n--- Try Your Own Sentence ---")
print("Type 'exit' to stop")

while True:
    user_input = input("\nEnter Hindi sentence: ")

    if user_input.lower() == 'exit':
        break

    tagged_output = tag_sentence(user_input)
    print("Tagged:", " ".join([f"{w}/{t}" for w, t in tagged_output]))
```
